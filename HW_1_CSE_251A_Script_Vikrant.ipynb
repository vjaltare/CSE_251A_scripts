# Question 5 Data Visualization

```python
## Loading the UCI wine dataset
wine = datasets.load_wine()

## loading data and target into X and Y variables
X = wine.data
Y = wine.target

# finding shapes of X and Y
print(f'shape of X: {X.shape}')
print(f'shape of Y: {Y.shape}')
```

```python
## plotting scatterplot for X and Y
fig, ax = plt.subplots(dpi=110, figsize=(3.5, 3.5))
scatter = ax.scatter(X[:, 0], X[:, 1], c = Y, alpha=.6)
# ax.scatter(X[:, 1], Y, label='feature 2', alpha=.6)
ax.set_xlabel('Feature 1: Alcohol', fontsize=14)
ax.set_ylabel('Feature 2: Malic Acid', fontsize=14)
ax.set_title('Scatter plot', fontsize=14)
handles, labels = scatter.legend_elements()
legend = ax.legend(handles, ['Wine 1', 'Wine 2', 'Wine 3'])
sns.despine()
fig.savefig('scatter_plot_hw1.pdf', bbox_inches='tight')
```
